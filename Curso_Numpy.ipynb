Portanto, utilizaremos o [arquivo raw](https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv) disponível no GitHub.

```python
d_dados = np.loadtxt('https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv', delimiter=',', usecols=np.arange(1,6,1), skiprows=1)
d_dados
```

As colunas que iremos avaliar são as de diâmetro e peso. Crie arrays específicos para guardar o diâmetro e peso da laranja e toranja. O diâmetro está na coluna zero e o peso na coluna 1. Os dados referentes a laranja vão até a linha 4999 e os referentes à toranja iniciam na linha 5000 do arquivo.

```python
diametro_laranja = d_dados[:5000,0]
peso_laranja = d_dados[:5000,1]
```

```python
diametro_toranja = d_dados[4999:,0]
peso_toranja = d_dados[4999:,1]
```

crie um gráfico para a laranja e para a toranja do peso pelo diâmetro.

```python
plt.plot(diametro_laranja,peso_laranja)
plt.plot(diametro_toranja,peso_toranja)
plt.legend(['laranja','toranja'])
plt.show()
```

 calcular o coeficiente ângular e o linear para a reta da laranja e para a reta da toranja.

n = Numero de elementos

X = diametro das frutas

Y = peso das frutas

a = coeficiente angular

b = coeficiente linear

###Laranja

```python
Y = peso_laranja
X = diametro_laranja
n = np.size(diametro_laranja)
```

```python
#coeficiente angular
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
a
```

```python
#coeficiente linear
b = np.mean(Y) - a*np.mean(X)
b
```

```python
y = a*X+b
```

```python
np.linalg.norm(diametro_laranja-y)
```

###Toranja

```python
Yt = peso_toranja
Xt = diametro_toranja
nt = np.size(diametro_toranja)
```

```python
#coeficiente angular
a = (nt*np.sum(Xt*Yt) - np.sum(Xt)*np.sum(Yt))/(n*np.sum(Xt**2) - np.sum(Xt)**2)
a
```

```python
#coeficiente linear
b = np.mean(Yt) - a*np.mean(Xt)
b
```

```python
yt = a*Xt+b
```

```python
np.linalg.norm(diametro_toranja-yt)
```
