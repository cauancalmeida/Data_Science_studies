```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
#criando bando de dados
engine = create_engine('sqlite:///:memory:')
```

```python
#Código omitido

url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine) #pegar indormações do BD
```

```python
print(inspector.get_table_names()) #retorna nome das tabelas
```

```python
query ='SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
empregados.to_sql('empregados', con = engine, index=False)
```

```python
pd.read_sql_table('empregados', con = engine)
```

```python
pd.read_sql_table('empregados', con = engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

###Desafio

```python
n_engine = create_engine('sqlite:///:memory:')
```

```python
dados_clientes = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv')
```

```python
dados_clientes.to_sql('clientes_new',n_engine, index=False)
```

```python
pd.read_sql_table('clientes_new',n_engine)
```

```python
def roda_query (query):
  #função para rodar todas as queries
  with n_engine.connect() as conn:
    return conn.execute(query)
```

```python
query_atualiza = 'UPDATE clientes_new SET Rendimento_anual = "300000" WHERE ID_Cliente = "6840104"'
roda_query(query_atualiza)
pd.read_sql_table('clientes_new',n_engine)
```

```python
query_deleta = 'DELETE FROM clientes_new WHERE ID_Cliente = "5008809"'
roda_query(query_deleta)
pd.read_sql_table('clientes_new',n_engine)
```

```python
query_insere = 'INSERT INTO clientes_new (ID_Cliente, Idade, Grau_escolaridade, Estado_civil, ' \
        'Tamanho_familia, Categoria_de_renda, Ocupacao, Anos_empregado, ' \
        'Rendimento_anual, Tem_carro, Moradia) ' \
        'VALUES (6850985, 33, "Doutorado", "Solteiro", 1, "Empregado", "TI", ' \
        '2, 290000, 0, "Casa/apartamento próprio")'
roda_query(query_insere)
pd.read_sql_table('clientes_new',n_engine)
```
