##Amostragem sistemática

Define um sistema para buscar os dados da amostra.

```python
dataset.shape
```

```python
len(dataset)//100
```

```python
random.seed(2)#semente de aleatoriedade
random.randint(0, 325)
```

```python
np.arange(28, len(dataset), step = 325)
```

```python
def amostragem_sistematica(dataset, amostras):
  intervalo = len(dataset) // amostras
  random.seed(2)
  inicio = random.randint(0, intervalo)
  indices = np.arange(inicio, len(dataset), step = intervalo)
  return dataset.iloc[indices]
```

```python
df_amostra_sistematica = amostragem_sistematica(dataset, 100)
df_amostra_sistematica
```

```python
def amostragem_agrupada(dataset, numero_grupos):
    intervalo = len(dataset) / numero_grupos
    grupos = []
    id_grupo = 0
    contagem = 0
    for index, row in dataset.iterrows():
        grupos.append(id_grupo)
        contagem += 1
        if contagem > intervalo:
            contagem = 0
            id_grupo += 1
    dataset['grupo'] = grupos
    random.seed(1)
    grupo_selecionado = random.randint(0, numero_grupos)
    return dataset[dataset['grupo'] == grupo_selecionado]
```

```python
df_amostra_agrupamento = amostragem_agrupada(dataset, 100)
df_amostra_agrupamento.shape
```

##Amostra estratificada

Vem de extrato, que vem de divisão.
A ideia é pegar uma amostra proporcional aos dados diferente.

Ex:
* População: 90 pessoas
* 54 mulheres e 36 homens
* Amostra de 10% da população: 9 pessoas
* Mulheres representam 60% da população ((54/90)*100)
* Homens representam 40% da população ((36/90)*100)
 * Quantidade de mulheres-> 54*10/100 = 5.4 (5)
 * Quantidade de homens -> 36*10/100 = 3.6 (4)

Isso é para saber quantos elementos de cada grupo devem ser selecionados. Pós calculo é necessario usar outra tecnica para seleção das amostras

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
dataset['income'].value_counts()
```

```python
7841 / len(dataset), 24720/len(dataset)
```

```python
split = StratifiedShuffleSplit(test_size=0.1) #0.1 = 10%
for x, y in split.split(dataset, dataset['income']):
  df_x = dataset.iloc[x] #90%
  df_y = dataset.iloc[y] #10%
```
