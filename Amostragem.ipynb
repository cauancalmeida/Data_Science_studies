##Amostragem sistemática

Define um sistema para buscar os dados da amostra.

```python
dataset.shape
```

```python
def amostragem_sistematica(dataset, amostras):
  intervalo = len(dataset) // amostras
  random.seed(1)
  inicio = random.randint(0, intervalo)
  indices = np.arange(inicio, len(dataset), step = intervalo)
  return dataset.iloc[indices]
```

```python
df_amostra_sistematica = amostragem_sistematica(dataset, 100)
df_amostra_sistematica
```

```python
def amostragem_agrupada(dataset, numero_grupos):
    intervalo = len(dataset) / numero_grupos
    grupos = []
    id_grupo = 0
    contagem = 0
    for index, row in dataset.iterrows():
        grupos.append(id_grupo)
        contagem += 1
        if contagem > intervalo:
            contagem = 0
            id_grupo += 1
    dataset['grupo'] = grupos
    random.seed(1)
    grupo_selecionado = random.randint(0, numero_grupos)
    return dataset[dataset['grupo'] == grupo_selecionado]
```

##Amostra estratificada

Vem de extrato, que vem de divisão.
A ideia é pegar uma amostra proporcional aos dados diferente.

Ex:
* População: 90 pessoas
* 54 mulheres e 36 homens
* Amostra de 10% da população: 9 pessoas
* Mulheres representam 60% da população ((54/90)*100)
* Homens representam 40% da população ((36/90)*100)
 * Quantidade de mulheres-> 54*10/100 = 5.4 (5)
 * Quantidade de homens -> 36*10/100 = 3.6 (4)

Isso é para saber quantos elementos de cada grupo devem ser selecionados. Pós calculo é necessario usar outra tecnica para seleção das amostras

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
dataset['income'].value_counts()
```

```python
7841 / len(dataset), 24720/len(dataset)
```

```python
split = StratifiedShuffleSplit(test_size=0.1) #0.1 = 10%
for x, y in split.split(dataset, dataset['income']):
  df_x = dataset.iloc[x] #90% -> base de treinamento
  df_y = dataset.iloc[y] #10% -> base de teste
```

```python
df_x.shape, df_y.shape
```

```python
def amostragem_estratificada(dataset, percentual):
  split = StratifiedShuffleSplit(test_size=percentual, random_state=1)
  for _, y in split.split(dataset, dataset['income']):
    df_y = dataset.iloc[y]
  return df_y
```

```python
df_amostra_estrat = amostragem_estratificada(dataset, 0.1)
df_amostra_estrat.shape
```

##Amostragem Reservatorio

Muito usado para  **Data stream(base de dados dinamica, ex.: compras em um e-commerce)** itens com tamanho desconhecido que pode ser acessado apenas uma vez.
Algoritimo para sortear um item do stream, porém, cada item deve possuir a mesma probabilidade de seleção.

```python
stream =[]
for i in range(len(dataset)):#lista de streams
  stream.append(i)
```

```python
def amostragem_reervatorio(dataset, amostras):
  stream = []
  for i in range(len(dataset)):#lista dos IDs dos stream vindos do dataset
    stream.append(i)
  i = 0
  reservatorio = [0] * amostras #Reserva os IDs dos registros selecionados
  for i in range(amostras):#inicia o reservatorio
    reservatorio[i] = stream[i]
  while i< len(dataset):
    j = random.randrange(i+1)
    if j < amostras:
      reservatorio[j] = stream[i] #reservatorio recebe os valores sorteados
    i += 1
  return dataset.iloc[reservatorio]
```

```python
df_amostragem_reservatorio = amostragem_reervatorio(dataset, 100)
df_amostragem_reservatorio.shape
```
