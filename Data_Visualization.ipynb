```python
import pandas as pd
dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv')
dados.head(10)
```

```python
dados.info()
```

#Tendencias de Imigração do Brasil

```python
dados.set_index('País', inplace=True)
dados.head()
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
brasil= dados.loc['Brasil', anos]
brasil
```

```python
brasil_dic={'ano': brasil.index.to_list(), 'imigrantes': brasil.values.tolist()}
brasil_df = pd.DataFrame(brasil_dic)
brasil_df
```

```python
import matplotlib.pyplot as plt
```

```python
#grafico de linhas é melhor para visualização de dados de acordo com o tempo
plt.figure(figsize=(10,5))
plt.plot(brasil_df['ano'], brasil_df['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Imigrantes do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.show()#depois do show, mais nada é executado
```

```python
argentina = dados.loc['Argentina', anos]
argentina
```

```python
argentina_dic={'ano': argentina.index.to_list(), 'imigrantes': argentina.values.tolist()}
argentina_df = pd.DataFrame(argentina_dic)
argentina_df
```

```python
plt.plot(brasil_df['ano'], brasil_df['imigrantes'])
plt.plot(argentina_df['ano'], argentina_df['imigrantes'])
plt.legend(['Brasil', 'Argentina'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Imigrantes do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.show()
```

##Há alguma tendência ou padrão comum nos dados dos dois países?

Entre os anos de 1985 e 1995 os dois paises tem um padrão de imigração para o Canadá muito parecido.

##Quais são os períodos com maior número de imigrantes nos dois países?

no Brasil o maior periodo de imigração foi entre 2005 e 2010 sendo mais especifico :

```python
maior_imigrates_br = brasil_df.sort_values('imigrantes', ascending=False).head(1)
maior_imigrates_br
```

Ja para a Argentina o maior periodo foi entre 2000 e 2005. Mais especificamente:

```python
maior_imigrates_ar = argentina_df.sort_values('imigrantes', ascending=False).head(1)
maior_imigrates_ar
```

```python
#fig é a figura a ser criada| ax vem de axes, são os eixos da figura
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(brasil_df['ano'], brasil_df['imigrantes'], label='Brasil')
ax.set_title('Imigrantes do Brasil e Argentina para o Canadá de\n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
#-------------- Linhas ------------------------------
fig, axs = plt.subplots(1,2, figsize=(15,8))
axs[0].plot(brasil_df['ano'], brasil_df['imigrantes'], label='Brasil')
axs[0].set_title('Imigrantes do Brasil para o Canadá de\n 1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
#------------- boxplot ----------------------------
axs[1].boxplot(brasil_df['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para o Canadá de\n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()
# --------------------------------------------------
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigrantes dos quatro maiores paises da America do Sul para o Canadá de\n 1980 a 2013')
#----------- Brasil ----------------
axs[0,0].plot(dados.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

#----------- Argentina ----------------
axs[0,1].plot(dados.loc['Argentina', anos])
axs[0,1].set_title('Argentina')

#----------- Colombia ----------------
axs[1,0].plot(dados.loc['Colômbia', anos])
axs[1,0].set_title('Colômbia')

#----------- Peru ----------------
axs[1,1].plot(dados.loc['Peru', anos])
axs[1,1].set_title('Peru')

#-----------------------------------
for ax in axs.flat:
  ax.set(xlabel='Ano', ylabel='Imigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()
ymin=0
ymax=7000
for ax in axs.ravel(): # revel -> consegue transformar um array bidimensional em unidimensional e fazer algumas aplicações
  ax.set_ylim(ymin, ymax)
plt.show()
```

##Desafio: visualizando dados de vendas de diferentes lojas

você deve criar quatro subplots dispostos em duas linhas e duas colunas, onde cada subplot representa uma loja diferente. Nesse desafio, cada subplot deve apresentar um gráfico de linhas que mostre a variação do número de vendas ao longo dos meses do ano.

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
vendas_2022 = pd.DataFrame(vendas_2022)
vendas_2022.index = lojas
vendas_2022
```

```python
#Criando a figura
fig, axs = plt.subplots(2,2, figsize=(10,6))
#ajustando espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.3)
#Adicionando titulo geral
fig.suptitle('Vendas de cada loja de 2022')
#-------------Loja A--------------------------
axs[0,0].plot(vendas_2022.loc['A'])
axs[0,0].set_title('Loja A')
#------------Loja B---------------------------
axs[0,1].plot(vendas_2022.loc['B'])
axs[0,1].set_title('Loja B')
#-----------Loja C----------------------------
axs[1,0].plot(vendas_2022.loc['C'])
axs[1,0].set_title('Loja C')
#----------Loja D------------------------------
axs[1,1].plot(vendas_2022.loc['D'])
axs[1,1].set_title('Loja D')
#----------------------------------------------
ymin=0
ymax=500
#adicionando label para os eixos
for ax in axs.flat:
  ax.set(xlabel='Mes', ylabel='Vendas')
  ax.grid()
#deixando todos os graficos na mesma escala no eixo y
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```
