```python
import pandas as pd
dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv')
dados.head()
```

```python
dados.info()
```

#Tendencias de Imigração do Brasil

```python
dados.set_index('País', inplace=True)
dados.head()
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
brasil= dados.loc['Brasil', anos]
brasil
```

```python
brasil_dic={'ano': brasil.index.to_list(), 'imigrantes': brasil.values.tolist()}
brasil_df = pd.DataFrame(brasil_dic)
brasil_df
```

```python
import matplotlib.pyplot as plt
```

```python
#grafico de linhas é melhor para visualização de dados de acordo com o tempo
plt.figure(figsize=(10,5))
plt.plot(brasil_df['ano'], brasil_df['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Imigrantes do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.show()#depois do show, mais nada é executado
```

```python
argentina = dados.loc['Argentina', anos]
argentina
```

```python
argentina_dic={'ano': argentina.index.to_list(), 'imigrantes': argentina.values.tolist()}
argentina_df = pd.DataFrame(argentina_dic)
argentina_df
```

```python
plt.plot(brasil_df['ano'], brasil_df['imigrantes'])
plt.plot(argentina_df['ano'], argentina_df['imigrantes'])
plt.legend(['Brasil', 'Argentina'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Imigrantes do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.show()
```

##Há alguma tendência ou padrão comum nos dados dos dois países?

Entre os anos de 1985 e 1995 os dois paises tem um padrão de imigração para o Canadá muito parecido.

##Quais são os períodos com maior número de imigrantes nos dois países?

no Brasil o maior periodo de imigração foi entre 2005 e 2010 sendo mais especifico :

```python
maior_imigrates_br = brasil_df.sort_values('imigrantes', ascending=False).head(1)
maior_imigrates_br
```

Ja para a Argentina o maior periodo foi entre 2000 e 2005. Mais especificamente:

```python
maior_imigrates_ar = argentina_df.sort_values('imigrantes', ascending=False).head(1)
maior_imigrates_ar
```
