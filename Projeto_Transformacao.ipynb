## Projeto desafio 1: vendas online



O objetivo desse projeto é realizar uma análise dos resultados de um evento com os clientes de uma empresa de vendas online. Foi coletado um conjunto de dados que contém os clientes que mais gastaram com produtos dentro de 5 dias de vendas, que é o período de duração do evento. Essa análise vai identificar o cliente com a maior compra na semana, que irá receber um prêmio da loja, e posteriormente, ela pode ajudar a empresa a criar novas estratégias para atrair mais clientes.

```python
dados_vendas = requests.get('https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json')
dados_vendas = pd.read_json(dados_vendas.text)
```

```python
dados_vendas.head()
```

```python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
dados_vendas
```

```python
#removendo os dados em lista do DF
cols = list(dados_vendas.columns)
dados_vendas = dados_vendas.explode(cols[1:])
dados_vendas
```

```python
#resetando index
dados_vendas.reset_index(drop=True, inplace=True)
dados_vendas
```

```python
dados_vendas.info()
```

```python
#limpando a coluna 'Valor da compra'
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('$','').replace(',','').replace('R','').strip())
dados_vendas['Valor da compra']
```

```python
#transformando coluna
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].astype(np.float64)
```

```python
dados_vendas.info()
```

```python
#tratando coluna cliente
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower()
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.replace('[^a-z]',' ',regex=True)
dados_vendas.head()
```

```python
#transformando coluna data de venda
dados_vendas['Data de venda']= pd.to_datetime(dados_vendas['Data de venda'],format='%d/%m/%Y')
dados_vendas.info()
```

```python
#Pegando total de vendas
total_compras = dados_vendas.groupby('Cliente')['Valor da compra'].sum()
total_compras
```

## Projeto desafio 2: administração de condomínios

A administração de condomínios é uma tarefa que requer muita atenção e organização. Entre as diversas responsabilidades da gestão está o recebimento dos aluguéis dos locatários. Para garantir uma boa saúde financeira do empreendimento, é fundamental que esses pagamentos sejam feitos de forma regular e pontual. No entanto, sabemos que nem sempre isso acontece.

Pensando nisso, proponho um desafio de tratamento de dados com o objetivo de analisar o atraso no pagamento de aluguéis no condomínio de alguns moradores. Disponibilizo a base de dados dados_locacao_imoveis.json, que contém informações sobre o apartamento dos locatários, o dia acordado para o pagamento do aluguel, o dia da efetivação do pagamento de aluguel e o valor do aluguel.

```python
dados_locacao = requests.get('https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json')
dados_locacao = pd.read_json(dados_locacao.text)
```

```python
dados_locacao.head()
```

```python
dados_locacao = pd.json_normalize(dados_locacao['dados_locacao'])
dados_locacao
```

```python
#removendo as listas
col_locacao = list(dados_locacao.columns)
dados_locacao = dados_locacao.explode(col_locacao[1:])
dados_locacao
```

```python
#resetando os index
dados_locacao.reset_index(drop=True, inplace=True)
dados_locacao
```

```python
#limpando coluna aluguel
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].apply(lambda x: x.replace('$','').replace(',','').replace('reais','').strip())
```

```python
dados_locacao.info()
```

```python
#transformando coluna valor_aluguel
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].astype(np.float64)
```

```python
dados_locacao.info()
```

```python
#tratando coluna apartamento
dados_locacao['apartamento'] = dados_locacao['apartamento'].str.replace('(blocoAP)','')
dados_locacao.head()
```

```python
#tratando as colunas de data
dados_locacao['datas_de_pagamento'] = pd.to_datetime(dados_locacao['datas_de_pagamento'],format='%d/%m/%Y')
dados_locacao['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao['datas_combinadas_pagamento'],format='%d/%m/%Y')
dados_locacao.info()
```

```python
#pegando o atrado
dados_locacao['atraso'] = (dados_locacao['datas_de_pagamento'] - dados_locacao['datas_combinadas_pagamento']).dt.days
dados_locacao
```

```python
#media de atraso
atraso_media = dados_locacao.groupby('apartamento')['atraso'].mean()
atraso_media
```
